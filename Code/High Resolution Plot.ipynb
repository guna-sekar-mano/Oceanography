```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import os
import xarray as xr
```

```python
chloro_path = 'C:/Users/Admin/Downloads/NOAA/2022'
sss_path = 'C:/Users/Admin/Downloads/Download SSS data from URL/SSS_New_Data'
sst_path = 'C:/Users/Admin/Downloads/SST/2022'
ssh_path = 'C:/Users/Admin/Downloads/AVISO/2022'

files_1 = [f for f in os.listdir(chloro_path) if f.endswith('.nc')]
files_2 = [f for f in os.listdir(sss_path) if f.endswith('.nc')]
files_3 = [f for f in os.listdir(sst_path) if f.endswith('.nc')]
files_4 = [f for f in os.listdir(ssh_path) if f.endswith('.nc')]
```

```python
ds_chloro = xr.open_mfdataset(os.path.join(chloro_path, f) for f in files_1)
ds_sss = xr.open_mfdataset(os.path.join(sss_path, f) for f in files_2)
ds_sst = xr.open_mfdataset(os.path.join(sst_path, f) for f in files_3)
ds_ssh = xr.open_mfdataset(os.path.join(ssh_path, f) for f in files_4)
```

```python
# Set up the projection and figure
proj = ccrs.PlateCarree()
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(projection=proj))

# Plot the variable of interest using imshow() method
ds_ssh.adt[1].interp(method='nearest').plot.imshow(ax=ax, transform=proj, cmap='jet', add_colorbar=True) #cubic and also use nearest

# Add coastlines and gridlines to the plot
ax.coastlines(resolution='50m')
ax.gridlines(linestyle='--', draw_labels=True)

# Set the title of the plot
plt.title('Sea Surface Height Anomaly using Nearest Method \n', fontsize=16)

# Show the plot
plt.show()
```
