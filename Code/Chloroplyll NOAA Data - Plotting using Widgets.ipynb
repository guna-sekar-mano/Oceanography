# Chloroplyll NOAA Data - Plotting using Widgets

```python
import warnings
warnings.filterwarnings('ignore')
import xarray as xr
import ipywidgets as widgets
import os
```

```python
path = 'C:/Users/Admin/Downloads/NOAA/2022'
files = [f for f in os.listdir(path) if f.endswith('.nc')]
```

```python
ds = xr.open_mfdataset(os.path.join(path, f) for f in files)

dates = ds['time'].values
```

```python
ds
```

```python
def plot_data(date):
    data = ds.sel(time=date)
    data['chlor_a'].plot()
```

```python
date_selector = widgets.Dropdown(options=dates, value=dates[0], description='Date:', layout={'width': 'initial'})
widgets.interact(plot_data, date=date_selector)
```
