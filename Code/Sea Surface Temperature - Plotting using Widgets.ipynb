# Sea Surface Temperature - Plotting using Widgets

```python
import warnings
warnings.filterwarnings('ignore')
import xarray as xr
import ipywidgets as widgets
```

```python
ds = xr.open_dataset("C:/Users/Admin/Downloads/SST/2022/SST_2022.nc", engine= "netcdf4")
```

```python
ds
```

```python
dates = ds['time'].values
```

```python
def plot_data(date):
    data = ds.sel(time=date)
    data['sea_surface_temperature'].plot()
```

```python
date_selector = widgets.Dropdown(options=dates, value=dates[0], description='Date:', layout={'width': 'initial'})
widgets.interact(plot_data, date=date_selector)
```
